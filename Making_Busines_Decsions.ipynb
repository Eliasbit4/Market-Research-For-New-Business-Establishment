### Apply the ICE framework to prioritize hypotheses. Sort them in descending order of priority.
<a href = "#Table-of-contents"> Back to table of contents</a>

Impact, confidence, effort/ease (ICE) is one of the most popular ways of prioritizing problems:

The below ICE SCORE formula is irrelevant since we do not have Ease, but we'll use the next one since we do have the Impact, Confidence, and Ease.

```python
hypotheses['ICE'] = hypotheses['impact']*hypotheses['confidence']/hypotheses['effort']
```

```python
hypotheses_sorted_ICE = hypotheses.sort_values(by='ICE', ascending=False)
hypotheses_sorted_ICE
```

```python
#https://stackoverflow.com/questions/43770507/seaborn-bar-plot-ordering
ax = sns.barplot("hypothesis", y="ICE", data=hypotheses_sorted_ICE,palette="Blues_d")
ax.set_xticklabels(ax.get_xticklabels(), rotation=85);
```

### Apply the RICE framework to prioritize hypotheses. Sort them in descending order of priority.
<a href = "#Table-of-contents"> Back to table of contents</a>

```python
hypotheses['RICE'] = hypotheses['reach']*hypotheses['impact']*hypotheses['confidence']/hypotheses['effort']
```

```python
hypotheses_sorted_RICE = hypotheses.sort_values(by='RICE', ascending=False)
hypotheses_sorted_RICE
```

```python
#https://stackoverflow.com/questions/43770507/seaborn-bar-plot-ordering
ax = sns.barplot("hypothesis", y="RICE", data=hypotheses_sorted_RICE,palette="Blues_d")
ax.set_xticklabels(ax.get_xticklabels(), rotation=85);
```

### Show how the prioritization of hypotheses changes when you use RICE instead of ICE . Provide an explanation for the changes.
<a href = "#Table-of-contents"> Back to table of contents</a>

As shown above, we applied the ICE framework to prioritize hypothesis and found that hypothesis 8 was the most promising one followed by hypothesis 0, 7, and 6.
We applied the RICE framework to prioritize hypotheses and found that hypothesis 7 was the most promising one followed by hypothesis 2, 0 and 6. The changes and differene of priotrization between the two framework is clear; it's related to the different formulas we used: 

ICE score = (Impact x Confidence) / Effort

VS

RICE score = (Reach x Impact x Confidence)/ Effort

As you can see above, in RICE, we multiplied by additional paramter which is the "Reach". Reach represents how many users will be affected by the update you want to introduce. Keep in mind, Impact, confidence, effort/ease (ICE) is one of the most popular ways of prioritizing problems. However, RICE is considered amodified version. Having said that, and assuming that the four components (Impact, Confidence, Reach, and Effort) numbers are accurate, it would be logical to use the RICE score to prioritizing problems.

See below the two graphs. It was obvious from the beggining to acually use RICE score to take a decsiaion since RICE is a modified version of ICE with more variables that help us take better decisions**

### Task 1. Graph cumulative revenue by group. Make conclusions and conjectures.
<a href = "#Table-of-contents"> Back to table of contents</a>

We'll build an array with unique paired date-group values using the drop_duplicates() method:

```python
# building an array with unique paired date-group values 
datesGroups = orders[['date','group']].drop_duplicates()
datesGroups.head()
```

Let's collect the aggregated cumulative daily data on orders in several steps.

In the first part of the code, we get the rows of the orders table where the date is less than or equal to the date of the datesGroups element and the test group is the same as in datesGroups:

orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].

* np.logical_and() is  function that's quite useful for calculating cumulative metrics is. It allows us to apply boolean operations to Series objects. This is helpful when you need to pick a subset of rows from a table that match several conditions.

In the second part, we aggregate the values. We find the maximum date. We'll find the maximum value for the group, too, so that we have the 'group' column in our final output. We find the number of unique order and user IDs and calculate the sum of average purchase sizes: 

.agg({'date' : 'max', 'group' : 'max', 'orderId' : pd.Series.nunique, 'userId' : pd.Series.nunique, 'revenue' : 'sum'})

In other words, with this code, it's as if we've been calculating the results of our test every day up to the current date and storing them in the rows of the ordersAggregated table.

* We get current date with maximum value. 

* The 'group' column could be 'A' or 'B', there's no mean value. We do it to keep the group in the output only (no calculations).

We'll apply the methods to each row of the DataFrame (apply(axix=1)) and sort the results by the columns 'date' and 'group'. Putting this all together, we get:

```python
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'transactionId' : pd.Series.nunique, 'visitorId' : pd.Series.nunique, 'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])
```

```python
ordersAggregated.head()
```

With this code, it's as if we've been calculating the results of our test every day up to the current date and storing them in the rows of the ordersAggregated table.

Let's do something similar to get the aggregated cumulative daily data on visitors:

```python
visitsAggregated = datesGroups.apply(lambda x: visits[np.logical_and(visits['date'] <= x['date'], visits['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'visits' : 'sum'}), axis=1).sort_values(by=['date','group'])
```

```python
visitsAggregated.head()
```

We merge the two tables into one and give its columns descriptive names:

```python
# merging the two tables into one and giving its columns descriptive names
cumulativeData = ordersAggregated.merge(visitsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visits']

cumulativeData.head(5)
```

Let's plot cumulative revenue graphs by day and A/B test group:

```python
# DataFrame with cumulative orders and cumulative revenue by day, group A
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]

# DataFrame with cumulative orders and cumulative revenue by day, group B
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

# Plotting the group A revenue graph 
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')

# Plotting the group B revenue graph 
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')

plt.legend()
plt.xticks(rotation=45, ha='right') # rotate x-axis labels by 45 degrees. https://stackoverflow.com/questions/10998621/rotate-axis-text-in-python-matplotlib

plt.xlabel('date')
plt.ylabel('cumulative revenue')
plt.title('Cumulative Revenue Graph')
```

Revenue is rising steadily throughout the whole test for group A. As of group B, in 2019-08-17 the revenue soared for some reason. That's a good sign for both groups.

As of group B 2019-08-17 revenue soaring, this may imply either a surge in the number of orders or the presence of very expensive orders in the sample.

### Task 9. Find the statistical significance of the difference in conversion between the groups using the raw data. Make conclusions and conjectures.
<a href = "#Table-of-contents"> Back to table of contents</a>

Let's calculate the statistical significance of the difference in conversion between the groups.

We'll create the variables ordersByUsersA and ordersByUsersB to store the columns ['userId', 'orders']. For users with at least one order, we'll indicate the number of orders made.

```python
ordersByUsersA = orders[orders['group']=='A'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersA.columns = ['visitorId', 'orders']

ordersByUsersB = orders[orders['group']=='B'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersB.columns = ['visitorId', 'orders']
```

Now we'll declare the variables sampleA and sampleB, with users from different groups and corresponding numbers of orders. Those who didn't order anything will have 0. This is necessary to prepare the samples for the Mann-Whitney test.


sampleA must have two parts:

1. A list of the number of orders for each user: ordersByUsersA['orders']. 
2. Zeroes for users who didn't order anything. Their number is the difference between the total number of visitors and the number of records on orders: 
    visitors[visitors['group']=='A']['visitors'].sum() - len(ordersByUsersA['orders'])

We'll create a pd.Series object with the required length:

 pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), name='orders')
 
 
We used the np.arange() function to create a list of indices. It works like the range() function but creates an array of indices in the np.array format required by pd.Series.

In pandas, functions are concatenated with pd.concat(). First we need to pass it what needs to be concatenated. In our case, it's the list from the first and second parts:

[ordersByUsersA['orders'],pd.Series(0, index=np.arange(visitors[visitors['group']=='A']['visitors'].sum() - len(ordersByUsersA['orders'])), name='orders')]

Then we'll pass the argument that specifies that Series objects are to be concatenated by row (in other words, to be written in sequence): pd.concat([...], axis=0). Here's what we get after doing the same with sampleB:

```python
sampleA = pd.concat([ordersByUsersA['orders'],pd.Series(0, index=np.arange(visits[visits['group']=='A']['visits'].sum() - len(ordersByUsersA['orders'])), name='orders')],axis=0)

sampleB = pd.concat([ordersByUsersB['orders'],pd.Series(0, index=np.arange(visits[visits['group']=='B']['visits'].sum() - len(ordersByUsersB['orders'])), name='orders')],axis=0)
```

We'll apply the criterion and format the p-value by rounding it to three decimal places.
Remember that sampleA stores a sample where each element is the number of orders made by a certain user, including zeroes. So the number of sampleA elements is the number of users, and the sum of all the elements is the number of orders. To get the group's conversion rate, we'll need to divide the sum of the orders by the number of users. Using the mean() method, we'll find the mean for sampleA. We'll find the conversion rate for group B in a similar way: sampleB.mean().

Let's print the relative conversion gain for group B: group B conversion / group A conversion - 1. We'll round the value to three decimal places.

```python
print("{0:.3f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))

print("{0:.3f}".format(sampleB.mean()/sampleA.mean()-1)) 
```

There is a 0.8% probability that I would observe a relative difference in conversion gain of .138 or more, (even if there was no real difference between groups A and B.), due to chance alone)

Conclusion: based on the raw data, there is a difference between groups A and B.

The first row of the output gives us the p-value, 0.008, which is less than 0.05. So we reject the null hypothesis that there's not a statistically significant difference in conversion between the groups. But the relative win of group B is 13.8% (the second row of the output).

**See below the explicit hypotheses testing for each of the tests**

```python
from scipy import stats as st

print('Sample size:', len(sampleA), len(sampleB))
print('Sample means: ', np.mean(sampleA), np.mean(sampleB))
print('Difference in mean:', np.mean(sampleA) - np.mean(sampleB))

alpha = .05 
difference_in_avg_conversion =  st.ttest_ind(sampleA, sampleB)

print('p-value: ', difference_in_avg_conversion.pvalue )
if (difference_in_avg_conversion.pvalue < alpha):
        print("We reject the null hypothesis: There is no difference in conversion between the groups using the raw data")
else:
        print("We can't reject the null hypothesis: There is a difference in conversion between the groups using the raw datay")
```
